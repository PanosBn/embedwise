```python
hf_encoder = HuggingFaceEncoder("bert-base-uncased")
```

```python
sent_encoder = SentenceEncoder()
```

```python
dataset = load_dataset("ag_news")
```

```python
texts = dataset['train']['text'][600:1000]
labels = dataset['train']['label'][600:1000]
```

```python
# Encode the texts
embeddings = sent_encoder.get_embedding(texts)

# Reduce dimensions
reduced_embeddings = sent_encoder.reduce_dimensions(embeddings, method='tsne')
```

```python
def map_label(num):
    labels = {0: "World", 1: "Sports", 2: "Business", 3: "Sci/Tech"}
    return labels.get(num, "Unknown")  # Default to "Unknown" if the number is not in the dictionary

text_labels = [map_label(num) for num in labels]
```

```python
import pandas as pd

# Create a DataFrame for the embeddings
df = pd.DataFrame(reduced_embeddings, columns=['x', 'y'])
df['label'] = text_labels  # Add the labels
df['text'] = texts
```

```python
import altair as alt


# Create a scatter plot
alt.Chart(df).mark_circle(size=60).encode(
    x='x',
    y='y',
    color='label:N',
    tooltip=['label:N', 'text']
).interactive()

```

```python
# Encode the texts
embeddings = hf_encoder.get_embedding(texts)

# Reduce dimensions
reduced_embeddings = hf_encoder.reduce_dimensions(embeddings, method='tsne')
```

```python
df_hf = pd.DataFrame(reduced_embeddings, columns=['x', 'y'])
df_hf['label'] = text_labels  # Add the labels
df_hf['text'] = texts
```

```python
alt.Chart(df_hf).mark_circle(size=60).encode(
    x='x',
    y='y',
    color='label:N',
    tooltip=['label:N', 'text']
).interactive()
```
